```python
#import pandas, matplotlib, and numpy

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#load data csv files

debt = pd.read_csv('Data/Debt_In_America_by_County_Sep2023.csv')
home = pd.read_csv('Data/Zillow_Home_Value_Index_ZHVI_by_County_Jan2023_to_Sep2023.csv')
rent = pd.read_csv('Data/Zillow_Observed_Rent_Index_ZORI_by_County_Jan2023_to_Sep2023.csv')
```

```python
debt.head()
```

```python
home.head()
```

```python
rent.head()
```

```python
#Merge dataframes on 'StateFIPS' and 'CountyFIPS'

# Merge 'debt' and 'home' first
merged_df = debt.merge(home, on=['StateFIPS', 'CountyFIPS'], how='inner', suffixes=('', '_home'))

# Then, merge the resulting dataframe with 'rent'
merged_df = merged_df.merge(rent, on=['StateFIPS', 'CountyFIPS'], how='inner', suffixes=('', '_rent'))
```

```python
merged_df.head()
```

```python
merged_df.columns
```

```python
merged_df.isnull().sum()
```

```python
# Create new columns 'Average Home Value' and 'Average Zillow Rent Estimate' that are the averages of monthly columns
merged_df['Average Zillow Home Estimate'] = merged_df[['1/31/23', '2/28/23', '3/31/23', '4/30/23','5/31/23','6/30/23','7/31/23','8/31/23','9/30/23']].mean(axis=1)
merged_df['Average Zillow Rent Estimate'] = merged_df[['1/31/23_rent', '2/28/23_rent', '3/31/23_rent', '4/30/23_rent','5/31/23_rent','6/30/23_rent','7/31/23_rent','8/31/23_rent','9/30/23_rent']].mean(axis=1)
print(merged_df)
```

```python
# Drop the monthly columns
merged_df = merged_df.drop(columns=['1/31/23', '2/28/23', '3/31/23', '4/30/23','5/31/23','6/30/23','7/31/23','8/31/23','9/30/23','1/31/23_rent', '2/28/23_rent', '3/31/23_rent', '4/30/23_rent','5/31/23_rent','6/30/23_rent','7/31/23_rent','8/31/23_rent','9/30/23_rent'])
merged_df.head()
```

```python
merged_df.isnull().sum()
```

```python
merged_df.shape
```

```python
# Drop 'FIPS' columns since only needed for merging
# Drop redundant columns

merged_df = merged_df.drop(columns=['StateFIPS','CountyFIPS','CombinedFIPS','County Name','SizeRank','RegionName','State','Metro','SizeRank_rent','Metro_rent'])
```

```python
merged_df.shape
```

```python
merged_df.isnull().sum()
```

```python
# Drop the last few rows with missing values
merged_df.dropna()
```

```python
# Rename Columns
merged_df = merged_df.rename(columns={
    'RegionName_rent': 'County',
    'State_rent': 'State',
    'Share with any debt in collections, All':'TotalDebt%',
    'Median debt in collections, All':'MedianDebtCollections',
    'Share with medical debt in collections, All':'MedicalDebt%',
    'Share of student loan holders with student loan debt in default, All':'StudentLoanDefault%',
    'Auto/retail loan delinquency rate, All':'AutoRetailLoanDelinquency%',
    'Credit card debt delinquency rate, All':'CreditCardDelinquency%',
    'Share of people of color':'PeopleOfColor%',
    'Average household income, All':'AverageHouseholdIncome',
    'Average household income, Comm of color':'AverageHouseholdIncomePOC',
    'Average household income, White comm':'AverageHouseholdIncomeWhite'
    
})
```

```python
merged_df.head()
```

```python
#Move 'County' and 'State' to columns 1 and 2

# Define the new column order
cols = ['County', 'State'] + [col for col in merged_df.columns if col not in ['County', 'State']]

# Reorder the columns
merged_df = merged_df[cols]

merged_df.columns
```

```python
print(merged_df.dtypes)
```

```python
merged_df.head()
```

```python
# Rounding Values

merged_df = merged_df.round({'TotalDebt%': 3, 'MedianDebtCollections': 0, 'MedicalDebt%': 3,
                             'StudentLoanDefault%': 3, 'AutoRetailLoanDelinquency%':3,
                            'AutoRetailLoanDelinquency%':3, 'CreditCardDelinquency%':3,
                             'PeopleOfColor%':3,'AverageHouseholdIncome':0,
                             'AverageHouseholdIncomePOC':0, 'AverageHouseholdIncomeWhite':0,
                             'Average Zillow Home Estimate':0,'Average Zillow Rent Estimate':0
                            })
merged_df.head()
```

```python
df_clean = merged_df
df_clean.to_csv('Data/data_clean.csv')
```
